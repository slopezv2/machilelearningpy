 ## AND -> &, OR -> | 

### AND -> &

```python
# Users which talk 300 minutes or more and they are from NY
data3 = data[(data["Day Mins"] > 300) & (data["State"] == "NY")]
data3
```

### Or Users those talk more than 300 minutes or they are from NY

```python
data4 = data[(data["Day Mins"] > 300) | (data["State"] == "NY")]
len(data4)
data4.shape
```

### Comparisons with > and <

```python
data5= data[ data["Day Mins"] < data["Night Mins"]]
```
