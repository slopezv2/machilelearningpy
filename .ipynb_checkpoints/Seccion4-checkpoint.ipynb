# Computo de los valores faltantes

```python
dataRellenado = data.fillna(0)
```

```python
dataRellenado
```

```python
dataRellenado = data.fillna("unknown")
```

```python
dataRellenado
```

```python
#Elegir para cada columna
data["body"].fillna(0)
```

```python
data["home.dest"].fillna("Unknown")
```

```python
#Rellenando con la media
data["age"] = data["age"].fillna(data["age"].mean())
```

```python
data["age"]
```

```python
#Rellenando usando el valor de adelante (el de arriba)
data["age"].fillna(method="ffill")
```

```python
#Rellenando usando el valor de atras (el de abajo)
data["age"].fillna(method="backfill")
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Customer Churn Model.txt')
```

```python
data.head()
```

```python
plt.hist(data["Day Calls"], bins = 20)
plt.xlabel("Nro de llamadas al da")
plt.ylabel("Frecuencia")
plt.title("Histograma del nro de llamadas al da")
```

```python
plt.hist(data["Day Calls"], bins = [0, 60, 120, 170, 180])
# Esto es definir uno mismo los intervalos que van juntos
#Esto se puede obtener con la frmula de sturges
```

```python
#Aplicando Sturges
import numpy as np
sturges = int(np.ceil(1 + np.log2(333)))
plt.hist(data["Day Calls"], bins= sturges )
```

# Boxplot

```python
plt.boxplot(data["Day Calls"])
plt.ylabel("Numero de llamadas diarias")
plt.title("Llamadas diarias")
```

```python
data["Day Calls"].describe()
```

```python
#Tamao intercuantlico - tamao de la caja:
q25 = data["Day Calls"].quantile(q=0.25)
q75 = data["Day Calls"].quantile(q=0.75)
box_size = q75 - q25
box_size
```

```python
#Tamao pivote:
```

```python
inf = data["Day Calls"].quantile(q=0.25) - 1.5 * box_size
sup = data["Day Calls"].quantile(q=0.75) + 1.5 * box_size
print(inf)
print(sup)
```
